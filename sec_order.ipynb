```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def main():  
    kp=2.0
    tau=3.0
    zeta=0.05
    dx=1.0
```

$\frac{Y(s)}{X(s)} = \frac{K_p}{\tau^2 s^2 + 2 \zeta \tau s + 1}$

$\tau^2 \frac{d^2y}{dt^2} + 2 \zeta \tau \frac{dy}{dt} = -y + K_p x$

```python
    #defining a function to return dy/dt
    def model(x,t):
        y = x[0]
        dy_dt = x[1]
        d2y_dt2 = (-2.0*zeta*tau*dy_dt - y + kp*dx)/tau**2
        return [dy_dt,d2y_dt2]
    
```

```python
    #defining t and y
    t = np.linspace(0,500,10000)
    u = odeint(model,[0,0],t)
    y = u[:,0]
```

```python
    #plotting the graph
    plt.figure(1)
    plt.plot(t,y,'b')
    plt.xlabel('Time')
    plt.ylabel('Response (y)')
    plt.show()
```

```python
main()
```
